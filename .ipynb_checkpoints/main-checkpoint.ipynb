```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# Reading whale returns
whale_return_csv = Path('Data/whale_returns.csv')
whale_return_dataframe=pd.read_csv(whale_return_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
# Count nulls
whale_return_dataframe.isnull().sum()
```

```python
# Drop nulls
whale_return_dataframe=whale_return_dataframe.dropna().copy()
whale_return_dataframe
```

```python
# Reading algorithmic returns
algo_return_csv = Path('Data/algo_returns.csv')
algo_return_dataframe=pd.read_csv(algo_return_csv,index_col="Date", parse_dates=True, infer_datetime_format=True)
# Count nulls
algo_return_dataframe.isnull().sum()
```

```python
# Drop nulls
algo_return_dataframe=algo_return_dataframe.dropna()
algo_return_dataframe
```

```python
# Reading S&P 500 Closing Prices
sp500_return_csv = Path('Data/sp500_history.csv')
sp500_return_dataframe=pd.read_csv(sp500_return_csv,index_col="Date", parse_dates=True, infer_datetime_format=True)
sp500_return_dataframe
```

```python
sp500_return_dataframe.dtypes
```

```python
# Clean identified numeric fields with $ symbol
sp500_return_dataframe["Close"] = sp500_return_dataframe["Close"].str.replace("$", "")
sp500_return_dataframe["Close"]
```

```python
# Convert `Close` from `object` to `float`
sp500_return_dataframe["Close"] = sp500_return_dataframe["Close"].astype("float")
# Set the date as the index to the DataFrame
#sp500_return_dataframe.set_index(pd.to_datetime(sp500_return_dataframe["Date"], infer_datetime_format=True), inplace=True)
sp500_return_dataframe.head()
```

```python
#Calculate Return
sp500_daily_returns = sp500_return_dataframe.pct_change()
sp500_daily_returns.dropna(inplace=True)
```

```python
sp500_daily_returns.head()
```

```python
# Rename Close column to "SP500 Return"
sp500_daily_returns = sp500_daily_returns.rename(columns={'Close':"SP500 Return"})
```

```python
#Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns
all_portfolios = pd.concat([whale_return_dataframe, algo_return_dataframe, sp500_daily_returns], axis="columns", join="inner")
all_portfolios.head()
```

```python
all_portfolios.tail()
```

```python
#Plotting daily returns of all portfolios
all_portfolios.plot(figsize=(20, 10), title="All portfolio daily returns");
```

```python
#Calculating and plotting cumulative returns for all portfolios.
cumulative_return=(1+all_portfolios).cumprod()
cumulative_return.plot(figsize=(20, 10), title="all cumulative returns");
```

```python
#Creating a box plot for each of the returns
all_portfolios.plot.box(figsize=(20, 10), title="All portfolio daily returns box plot");
```

```python
#Calculate the standard deviation for each portfolio
all_portfolios.std()
```

```python
#Determine which portfolios are riskier than the S&P 500.
SP500_std = all_portfolios["SP500 Return"].std()
SP500_std
```

```python
#Comparing all_portfolios.std() vs SP500_std (True means riskier than SP500)
all_portfolios.std() > SP500_std
```

```python
#Calculating the Annualized Standard Deviation
annual_std = all_portfolios.std() * np.sqrt(252) 
annual_std
```

```python
#Calculating and plotting the rolling standard deviation for all portfolios using a 21-day window
all_portfolios.rolling(window=21).std().plot(figsize=(20,10), title="21 day std of all portfolios");
```

```python
#Calculating and plotting the correlation between each stock to determine which portfolios may mimick the S&P 500
all_portfolios_correlation = all_portfolios.corr()
```

```python
all_portfolios_correlation.style.background_gradient(cmap="rainbow")
```

```python
#Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500.
rolling_covariance = all_portfolios["SOROS FUND MANAGEMENT LLC"].rolling(window=60).cov(all_portfolios['SP500 Return'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 60-Day Covariance of SOROS FUND MANAGEMENT LLC Returns vs. S&P 500 Returns');
```

```python
rolling_variance = all_portfolios['SP500 Return'].rolling(window=60).var()
rolling_variance.plot(figsize=(20, 10), title='Rolling 60-Day Variance of S&P 500 Returns');
```

```python
# Calculate 30-day rolling beta of SOROS FUND MGMT LLC and plot the data
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of SOROS FUND MANAGEMENT LLC');
```

```python
# Calculating exponentially weighted moving average
all_portfolios.ewm(21).std().plot(figsize=(20,10));
```

```python
# Calculating and plotting Sharpe Ratio
sharpe_ratios = (all_portfolios.mean() * 252) / (all_portfolios.std() * np.sqrt(252))
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios");
```

```python
# Path the stock return csv
TSLA_return_csv = Path('Data/Crypto Daily Return Data - TSLA Return.csv')
NFLX_return_csv = Path('Data/Crypto Daily Return Data - NFLX Return.csv')
FB_return_csv = Path('Data/Crypto Daily Return Data - FB Return.csv')
```

```python
# Reading in csv and assigning variables
TSLA_return_df=pd.read_csv(TSLA_return_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
NFLX_return_df=pd.read_csv(NFLX_return_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
FB_return_df=pd.read_csv(FB_return_csv, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
TSLA_return_df=TSLA_return_df.rename(columns={'Close': "TSLA Close"})
NFLX_return_df=NFLX_return_df.rename(columns={'Close': "NFLX Close"})
FB_return_df=FB_return_df.rename(columns={'Close': "FB Close"})
```

```python
# Concat hand picked stocks daily returns
all_portfolios_2 = pd.concat([TSLA_return_df, NFLX_return_df, FB_return_df], axis="columns", join="inner")
all_portfolios_2.head()
```

```python
# Calculating daily returns of my custom portfolio
all_portfolios_2_daily_returns = all_portfolios_2.pct_change()
all_portfolios_2_daily_returns = all_portfolios_2_daily_returns.dropna()
all_portfolios_2_daily_returns
```

```python
# Calculating weighted returns
weights = [1/3, 1/3, 1/3]
portfolio_returns = all_portfolios_2_daily_returns.dot(weights)
portfolio_returns.tail()
```

```python
all_portfolios['Custom'] = portfolio_returns
all_portfolios.head()
```

```python
#Checking first few rows of consolidated portfolio
all_portfolios.head()
```

```python
#Dropping null
all_portfolios = all_portfolios.dropna().copy()
```

```python
all_portfolios.head()
```

```python
#Annualized Standard Deviation
risk = all_portfolios.std()*np.sqrt(252)
risk
```

```python
#Calculating and plotting 21 rolling days std
all_portfolios.rolling(window=21).std().plot(figsize=(20,10), title="21 day std of all portfolios");
```

```python
#Calculating and plotting the correlation between each stock to determine which portfolios may mimick the S&P 500
all_portfolios_correlation = all_portfolios.corr()
all_portfolios_correlation.style.background_gradient(cmap="rainbow")
```

```python
#Calculating and plotting the 60-day rolling beta between it and the S&P 500.
rolling_covariance = all_portfolios["Custom"].rolling(window=60).cov(all_portfolios['SP500 Return'])
rolling_variance = all_portfolios['SP500 Return'].rolling(window=60).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of Custom Portfolio');
```

```python
# Calculating and plotting Sharpe Ratios
sharpe_ratios = (all_portfolios.mean() * 252) / (all_portfolios.std() * np.sqrt(252))
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios");
```
